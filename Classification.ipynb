```
from __future__ import absolute_import, division, print_function, unicode_literals
import tensorflow as tf
import numpy as np
import pandas as pd

CSV_COLUMN_NAMES = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', 'Species']
SPECIES = ['Setosa', 'Versicolor', 'Virginica']

#getting data and saving into file
train_path=tf.keras.utils.get_file("iris_training.csv","https://storage.googleapis.com/download.tensorflow.org/data/iris_training.csv")
test_path=tf.keras.utils.get_file("iris_test.csv", "https://storage.googleapis.com/download.tensorflow.org/data/iris_test.csv")

#reading the frm file paths using pandas
train_data=pd.read_csv(train_path,names=CSV_COLUMN_NAMES, header=0)
test_data=pd.read_csv(test_path,names=CSV_COLUMN_NAMES, header=0)

#grab the datasets and read them into panda dataframe

print(train_data.shape)

#pop the species col and use it as our lable
train_y=train_data.pop('Species')
test_y=test_data.pop('Species')
print(train_y.head())

#make input function
def input_fn(traindata,labels,training=True,batch_size=256):
  #convert daatframes into datasets
  ds=tf.data.Dataset.from_tensor_slices((dict(traindata),labels))

  #shuffling
  if training:
    ds=ds.shuffle(1000).repeat()
  
  return ds.batch(batch_size)

  #create feature columns
feature_col=[]
for key in train_data.keys():
  feature_col.append(tf.feature_column.numeric_column(key=key))
print(feature_col)

#create model
#DNNClassifier
model_classifier=tf.estimator.DNNClassifier(feature_columns=feature_col,hidden_units=[40,15],n_classes=3)

#training a model with our training data
model_classifier.train(input_fn=lambda:input_fn(train_data,train_y,training=True),steps=5000)

#evaluate with testing data
result = model_classifier.evaluate(
    input_fn=lambda: input_fn(test_data, test_y, training=False))

print('\nTest set accuracy: {accuracy:0.3f}\n'.format(**result))
```
